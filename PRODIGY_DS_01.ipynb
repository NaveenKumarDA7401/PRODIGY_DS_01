Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.

HISTOGRAM

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the main population dataset
file_path = r"C:\Users\Lenovo\Downloads\API_SP.POP.TOTL_DS2_en_csv_v2_87\API_SP.POP.TOTL_DS2_en_csv_v2_87.csv"
metadata_file = r"C:\Users\Lenovo\Downloads\API_SP.POP.TOTL_DS2_en_csv_v2_87\Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_87.csv"
```

```python
# Load the dataset, skipping metadata rows
df = pd.read_csv(file_path, skiprows=4)
```

```python
# Select relevant columns (Country Name and population for the latest available year, 2023)
df_population = df[['Country Name', '2023']].dropna()
```

```python
# Convert population to numeric values
df_population['2023'] = pd.to_numeric(df_population['2023'], errors='coerce')
```

```python
# Load country metadata to identify actual countries
metadata = pd.read_csv(metadata_file)
```

```python
# Extract the list of actual countries (excluding regions or groups)
actual_countries = metadata[metadata['Region'].notna()]['TableName'].tolist()
```

```python
# Filter the population dataset to include only actual countries
df_population = df_population[df_population['Country Name'].isin(actual_countries)]
```

```python
# Plot the histogram of population distribution
plt.figure(figsize=(12, 6))
plt.hist(df_population['2023'], bins=30, edgecolor='black', alpha=0.7)
plt.xlabel("Total Population (log scale)")
plt.ylabel("Number of Countries")
plt.yscale("log")  # Log scale for better visualization of distribution
plt.title("Distribution of Total Population Across Countries (2023)")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

BAR GRAPH

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the main population dataset
file_path = r"C:\Users\Lenovo\Downloads\API_SP.POP.TOTL_DS2_en_csv_v2_87\API_SP.POP.TOTL_DS2_en_csv_v2_87.csv"
metadata_file = r"C:\Users\Lenovo\Downloads\API_SP.POP.TOTL_DS2_en_csv_v2_87\Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_87.csv"
```

```python
# Load the dataset, skipping metadata rows
df = pd.read_csv(file_path, skiprows=4)
```

```python
# Select relevant columns (Country Name and population for the latest available year, 2023)
df_population = df[['Country Name', '2023']].dropna()
```

```python
# Convert population to numeric values
df_population['2023'] = pd.to_numeric(df_population['2023'], errors='coerce')
```

```python
# Load country metadata to identify actual countries
metadata = pd.read_csv(metadata_file)
```

```python
# Extract the list of actual countries (excluding regions or groups)
actual_countries = metadata[metadata['Region'].notna()]['TableName'].tolist()
```

```python
# Filter the population dataset to include only actual countries
df_population = df_population[df_population['Country Name'].isin(actual_countries)]
```

```python
# Sort by population for better visualization
df_population = df_population.sort_values(by='2023', ascending=False).head(20)
```

```python
# Plot the bar chart of population distribution
plt.figure(figsize=(12, 6))
plt.bar(df_population['Country Name'], df_population['2023'], color='skyblue', edgecolor='black')
plt.xlabel("Country")
plt.ylabel("Total Population")
plt.title("Top 20 Most Populated Countries in 2023")
plt.xticks(rotation=90)  # Rotate country names for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

Observations from the Histogram and Bar Chart


Histogram (Distribution of Total Population Across Countries - 2023)
Skewed Distribution

The histogram likely shows a right-skewed (positively skewed) distribution, meaning most countries have relatively small populations, while a few have very large populations.
Log Scale Importance

Using a logarithmic scale helps visualize differences across countries because the population sizes vary dramatically (e.g., small island nations vs. populous countries like India and China).
Majority of Countries Have Moderate Populations

Most countries have populations in the tens of millions rather than in the billions, as only a few exceed hundreds of millions.
Outliers

A small number of countries (e.g., China, India, USA) stand out as extreme values with significantly larger populations compared to others.
Bar Chart (Top 20 Most Populated Countries - 2023)
Largest Populations

The bar chart highlights the top 20 most populated countries, showing China and India leading by a significant margin.
Steep Drop-Off in Population

After the first few countries (China, India, USA), there is a noticeable drop in population size, illustrating a steep decline after the top nations.
Regional Trends

The top 20 most populated countries include nations from Asia, Africa, and the Americas, indicating where the world's largest populations are concentrated.
Close Competition

Some countries (e.g., Indonesia, Pakistan, Brazil) have similar population sizes, showing clusters of nations with comparable demographic weights.
Country-Specific Insights

The ranking and relative heights of bars help compare countries directly, making it easier to identify where a nation stands globally in terms of population.


Conclusion

Histogram gives a broad view of how populations are distributed across all countries, showing general trends and outliers.
Bar Chart focuses on specific countries, allowing direct comparisons and highlighting the most populated nations.
